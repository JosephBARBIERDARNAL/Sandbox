```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.random.rand(50)
y = np.random.rand(50)

# Create the plot
fig, ax = plt.subplots()
ax.scatter(x, y,  s=50)

# Add a title and labels
ax.set_title('Scatter Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection

# Create some data
x = np.random.rand(50)
y = np.random.rand(50)

# Create the plot
fig, ax = plt.subplots()

# Create a list of circles
circles = []
for i in range(len(x)):
    circle = Circle((x[i], y[i]), 0.01)
    circles.append(circle)

# Create a PatchCollection from the list of circles
patch_collection = PatchCollection(circles, match_original=True)

# Add the PatchCollection to the axes
ax.add_collection(patch_collection)

# Set the limits of the plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Add a title and labels
ax.set_title('Scatter Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Show the plot
plt.show()
```
