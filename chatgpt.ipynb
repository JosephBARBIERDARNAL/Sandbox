```python
from openai import OpenAI

api_key = "sk-aaaaaaaaaaaa" # A REMPLACER (attention: à ne pas partager)
client = OpenAI(api_key=api_key)

def call_gpt(client, prompt, pre_prompt, model="gpt-3.5-turbo", temperature=0.1):

    completion = client.chat.completions.create(
        model=model,
        messages=[{"role": "system", "content": pre_prompt},
                  {"role": "user", "content": prompt}],
                  temperature=temperature)
    return completion.choices[0].message.content

# Exemple d'utilisation
prompt = "Quelle est la capitale de la France ?"
pre_prompt = "Répond aux question avec de l'humour"
call_gpt(client, prompt, pre_prompt)
```
