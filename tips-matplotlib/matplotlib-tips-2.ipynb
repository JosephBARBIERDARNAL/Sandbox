## Remove Right and Top Spines

Eliminating the top and right spines from a plot can significantly improve its visual appeal and clarity. This modification is particularly effective for **scatter**, **bubble**, and **time series** charts, where these spines may be visually intrusive and unnecessary.

This adjustment directs the viewer’s focus towards the data itself, creating a more streamlined and professional presentation by minimizing the plot’s frame.

### How to use:

```python
ax.spines[['top', 'right']].set_visible(False)
```

````python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['savefig.dpi'] = 300

fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y, 'o-', color='darkred', label='sin(x)')
plt.savefig('with-spines.png')


fig, ax = plt.subplots()
ax.spines[['top', 'right']].set_visible(False)
ax.plot(x, y, 'o-', color='darkred', label='sin(x)')
plt.savefig('without-spines.png')
````
