## Change coordinate system in a plot

By default, when you add an annotation in matplotlib, the coordinates are in data coordinates. For example if your data range from 10 to 30 on the x-axis, and you want to add an annotation at x=20, you would use the following code:

```python
fig, ax = plt.subplots()
ax.plot([10, 30], [0, 10])
ax.text(x=20, y=4, s='Annotation')
plt.show()
```

But the problem with that is if you want to add an element that should be independent of the data, like a title, your code would not be consistent. Fortunately, matplotlib provides a `transform` argument in its `text()` function that allows you to change the coordinate system. 

For example, if you want to add an annotation where the position is relative to the whole figure, you can use the `fig.transFigure` transform:

```python
fig, ax = plt.subplots()
ax.plot([10, 30], [0, 10])
fig.text(x=0.5, y=0.4, s='Annotation', transform=fig.transFigure)
plt.show()
```

Actually, there are lots of other possibilities, and this is extensively explained in a [dedicated post in the Python Graph Gallery](https://python-graph-gallery/589-how-to-change-coordinate-system)

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot([10, 30], [0, 10])
ax.text(x=20, y=4, s='Annotation')
plt.savefig('ax_data.png', dpi=300, bbox_inches='tight')
plt.show()
````

````python
fig, ax = plt.subplots()
ax.plot([10, 30], [0, 10])
fig.text(x=0.5, y=0.4, s='Annotation', transform=fig.transFigure)
plt.savefig('fig_data.png', dpi=300, bbox_inches='tight')
plt.show()
````
