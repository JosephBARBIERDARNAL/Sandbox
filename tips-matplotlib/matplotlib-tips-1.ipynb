## DPI Argument in Matplotlib

### Discovering the Magic of DPI

**DPI (dots per inch)** is the key to **stunning visual clarity** in your Matplotlib images. Higher DPI means **crisper, more detailed visuals**, perfect for **print and high-resolution displays**. Conversely, lower DPI offers **quick-loading images** ideal for web use.

### How to Set DPI in Matplotlib
x
Achieve **consistent high-quality output** by setting DPI globally for all your figures with the `rcParams` dictionary:

```python
import matplotlib.pyplot as plt

# Set global DPI for all figures
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

This approach ensures every figure you create is **beautifully detailed**, balancing **image quality, performance, and storage** perfectly.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, 'o-', color='darkred', label='sin(x)')

plt.rcParams['savefig.dpi'] = 300
plt.savefig('sin-highrez.png')

plt.rcParams['savefig.dpi'] = 70
plt.savefig('sin-lowrez.png')
````
