```python
import pandas as pd
import matplotlib.pyplot as plt
from matoolkit.draw_arrow import draw_arrow

fig, ax = plt.subplots()
ax.scatter(
   [1, 2, 3, 4, 5],
   [1, 2, 10, 4, 5],
)
draw_arrow(
   tail_position=(0.2, 0.3),
   head_position=(0.52, 0.8),
   fig=fig,
   color='skyblue'
)

plt.savefig('draw_arrow_1.png', bbox_inches='tight', dpi=300)
plt.show()
```
