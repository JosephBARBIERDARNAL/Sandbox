```python
import numpy as np
import imageio
import os
from PIL import Image

# Set the directory where your images are stored and the output size
images_folder = 'img'
output_size = (2998, 2880)
images_filenames = ['map'+str(i)+'.png' for i in range(2, 6)]

# Prepare the list for the images to be saved into the GIF
images_for_gif = []

for filename in images_filenames:
    with Image.open(os.path.join(images_folder, filename)) as img:
        # Calculate the cropping box
        left = (img.width - output_size[0]) / 2
        top = (img.height - output_size[1]) / 2
        right = (img.width + output_size[0]) / 2
        bottom = (img.height + output_size[1]) / 2

        # Crop and convert the image to a numpy array
        img_cropped = img.crop((left, top, right, bottom))
        img_array = np.array(img_cropped)

        # Append the numpy array to the list
        images_for_gif.append(img_array)

# Specify the output filename and duration for each frame (in milliseconds)
imageio.mimsave('output.gif', images_for_gif, duration=400, format='GIF', loop=10)

print('GIF created!')
```
